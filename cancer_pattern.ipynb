```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_url = "C:/Users/roywa/PycharmProjects/pattern_analysis_us/data/500_Cities__Cancer.csv"
states_url = "C:/Users/roywa/PycharmProjects/pattern_analysis_us/data/US_State_Boundaries/US_State_Boundaries.shp"
cancer_df = pd.read_csv(file_url)
states_gdf = gpd.read_file(states_url)
```

```python
cancer_gdf = gpd.GeoDataFrame(cancer_df, geometry=gpd.points_from_xy(cancer_df["Lon"], cancer_df["Lat"]))
```

```python
cancer_gdf = cancer_gdf[cancer_gdf.Lat.between(24, 50) & cancer_gdf.Lon.between(-125, -66)]
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
cancer_gdf.plot(ax=ax, color="red")
states_gdf.boundary.plot(ax=ax, color="black")
plt.show()
```
